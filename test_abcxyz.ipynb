```python
from scipy.stats import gmean
def Nguong(cong_thuc):
    Temp_ex = vis.explain_f(cong_thuc)

    temp_x= Temp_ex['Out_Value']
    # print(temp_x)
    Geo_Nguong = vis.get_formula_geomean_profit(cong_thuc)

    x_nguong = min(temp_x) - 10**(-10)
    # print('iii',Geo_Nguong,x_nguong)
    for x in temp_x:
        Geo_= []
        for i in range(0,len(Temp_ex['Out_Value'])):
            if Temp_ex['Out_Value'][i] > x:
                Geo_.append(Temp_ex['Out_profit'][i])
            else:
                Geo_.append(1.01)
        # print(gmean(Geo_),x)
        if gmean(Geo_) > Geo_Nguong:
            # print(gmean(Geo_),Geo_)
            Geo_Nguong = gmean(Geo_)
            x_nguong = x
        
        print(Geo_Nguong, x_nguong, x)

    return Geo_Nguong,x_nguong

Nguong("+56*65-27")
```

```python
import numpy as np
from numba import njit
@njit
def weighted_random(arr):
    """Trả ra index"""
    sum_ = np.sum(arr)
    if sum_ <= 1e-15:
        raise

    rand_ = np.random.uniform(0, sum_)

    for i in range(arr.shape[0]):
        rand_ -= arr[i]
        if rand_ <= 0:
            return i


from collections import Counter
```
